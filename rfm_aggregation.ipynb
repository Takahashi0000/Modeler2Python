# RFM 集計ノード

```python
import pandas as pd
import datetime
```

```python
df_tran = pd.read_csv("./data/sampletranDEPT4en2019.csv")
df_tran.head()
```

```python
# 日付をdatetimeで取り扱う
df_tran["SDATE_datetime"]=df_tran["SDATE"].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d'))

# 購買履歴データ内の最新日時を取得し、Recencyの基準日とする
NOW=df_tran["SDATE_datetime"].max()

# Recency,Frequency,Monetaryを集計する
df_rfm=df_tran.groupby("CUSTID").agg({"SDATE_datetime":lambda date: (NOW-date.max()).days,"SDATE":lambda date: len(date),"SUBTOTAL":lambda price:price.sum()}).reset_index()

# カラム名を変更
df_rfm=df_rfm.rename(columns={"SDATE_datetime":"Recency","SDATE":"Frequency","SUBTOTAL":"Monetary"})
```
